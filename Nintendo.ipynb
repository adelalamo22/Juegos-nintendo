```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd 
import requests
import unicodedata
```

```python
options=webdriver.ChromeOptions()
options.add_argument('--incognito')

driver = webdriver.Chrome(executable_path='./chromedriver.exe', options=options)

```

```python
busqueda='juegos+nintendo+switch'
url='https://www.amazon.es/s?k='
driver.get(url+busqueda)
```

```python
content=driver.page_source
soup=BeautifulSoup(content,'html.parser')
```

```python
soup
```

```python
sopup_nombre=soup.findAll(class_='a-size-mini a-spacing-none a-color-base s-line-clamp-4')
```

```python
sopup_precio=soup.findAll(class_='a-price-whole')
```

```python
lista_nombres=[]
for nombre in sopup_nombre:
    lista_nombres.append(nombre.text)
```

```python
lista_precios=[]
for precio in sopup_precio:
    lista_precios.append(precio.text)
```

```python
juntos=list(zip(lista_nombres, lista_precios))
```

```python
juntos
```

```python
df_lista=pd.DataFrame(juntos)
```

```python
df_lista.rename(columns={0:'juego',1:'precio'},inplace=True)
```

```python
df_lista.info()
```

```python
df_lista['precio']=df_lista['precio'].str.replace(',' , '.')
```

```python
df_lista['precio']=df_lista['precio'].astype(float)
```

```python
df_lista.sort_values(by='precio',ascending=False)
```

```python
df_lista['juego']=df_lista['juego'].str.replace('\n\n' , '')
```

```python
df_lista
```

```python
df_lista.to_html('juegos_amazon.html')
```

```python
df_lista.to_csv('juegos_amazon.csv',index=False)
```
